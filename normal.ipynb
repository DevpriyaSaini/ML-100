```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['class label','Alcohol','Malic Acid']
```

```python
sns.kdeplot(df['Alcohol'],shade=True)
```

```python
sns.kdeplot(df['Malic Acid'],shade=True)
```

```python
color_dsit={1:'red',2:'blue',3:'green'}
sns.scatterplot(x='Alcohol',y='Malic Acid',hue='class label',data=df,palette=color_dsit)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('class label',axis=1),df['class label'],test_size=0.3,random_state=0)
```

```python
x_test
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Ensure x_train and x_test are DataFrames
x_train = pd.DataFrame(x_train, columns=['Alcohol', 'Malic Acid'])  # example names
x_test = pd.DataFrame(x_test, columns=['Alcohol', 'Malic Acid'])

# Scale
scaler = MinMaxScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale  = scaler.transform(x_test)

# Convert scaled arrays back into DataFrames
x_train_scale = pd.DataFrame(x_train_scale, columns=x_train.columns)
x_test_scale  = pd.DataFrame(x_test_scale, columns=x_test.columns)
```
