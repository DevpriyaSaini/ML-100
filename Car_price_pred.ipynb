```python
# year has many non year values
#year is in objectsnot in int
#Price col ask for price and also in obj
#kms-drive col has kms with them and obj to int and nan value remove
#fuel type has nan value
#keep words of name
```

```python
 backup=df.copy()
```

```python
df=df[df['year'].str.isnumeric()]
```

```python
df.shape
```

```python
df['year']=df['year'].apply(lambda x:int(x))
```

```python
df.info()
```

```python
df=df[df['Price']!='Ask For Price']
```

```python
df.shape
```

```python
df['Price']=df['Price'].str.replace(',','').astype(int)
```

```python
df['kms_driven']=df['kms_driven'].str.split(" ").str.get(0).str.replace(',','')
```

```python
df=df[df['kms_driven'].str.isnumeric()]
```

```python
df.shape
```

```python
df=df[~df['fuel_type'].isna()]
```

```python
df['name']=df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df.head()
```

```python
df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df=df[df['Price']<6e6].reset_index(drop=True)
```

```python
df.to_csv('Cleaned_Car_data.csv')
```

model building

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns='Price'),df['Price'],test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(df[['name','company','fuel_type']])
```

```python
column_trns=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python

pipe=make_pipeline(column_trns,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
  X_train,X_test,y_train,y_test=train_test_split(df.drop(columns='Price'),df['Price'],test_size=0.2,random_state=i)
  lr=LinearRegression()
  pipe=make_pipeline(column_trns,lr)
  pipe.fit(X_train,y_train)
  y_pred=pipe.predict(X_test)
  scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[433]
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns='Price'),df['Price'],test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trns,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
