```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=5,n_classes=2,random_state=42)
```

```python
df=pd.DataFrame(X,columns=['f1','f2','f3','f4','f5'])
df['target']=y
print(df.shape)
df.head()
```

```python
#choose the row
def sample_row(df,per):
  return df.sample(int(per*df.shape[0]),replace=True)
```

```python
#choose the col
def sample_col(df,per):
  return df.iloc[:,np.random.choice(df.shape[1],int(per*df.shape[1]),replace=False)]
```

```python
df1=sample_row(df,0.2)
df2=sample_row(df,0.2)
df3=sample_row(df,0.2)
```

```python
df1
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
clf1=DecisionTreeClassifier()
clf2=DecisionTreeClassifier()
clf3=DecisionTreeClassifier()
```

```python
clf1.fit(df1.iloc[:,:-1],df1['target'])
clf2.fit(df2.iloc[:,:-1],df2['target'])
clf3.fit(df3.iloc[:,:-1],df3['target'])
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```
